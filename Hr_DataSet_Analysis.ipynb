```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"C:\Users\hazar\OneDrive\Desktop\Data sets For Practice\HR Data.csv")
data
```

### Removing unnecessary columns

```python
Cols_Remove = ['EmployeeCount','StandardHours','EmployeeNumber']
data_cleaned = data.drop(columns = Cols_Remove)
data_cleaned
```

### Renaming the columns

```python
Cols_Rename = {
              'Age':'Emp_Age',
              'Attrition':'Emp_Attrition',
              'BusinessTravel':'Business_Travel_Frequency',
              'DailyRate':'Daily_Rate',
              'DistanceFromHome':'Distance_From_Home',
              'Education':'Education_Level',
              'EducationField':'Field_of_Education',
              'TotalWorkingYears':'Total_Working_Years',
              'YearsAtCompany':'Years_At_Company',
              'YearsInCurrentRole':'Years_In_Current_Role'
              }
data_cleaned.rename(columns = Cols_Rename,inplace = True)
data_cleaned
```

### Eliminating redundant entries.

```python
data_cleaned.drop_duplicates(inplace = True)
data_cleaned
```

### Sanitizing specific columns.


```python
data_cleaned['Emp_Attrition']= data_cleaned['Emp_Attrition'].str.capitalize()
data_cleaned
```

### Eliminating the NaN values.


```python
data_cleaned.dropna(inplace = True)
data_cleaned
```

```python
Cleaned_Dataset = data_cleaned
Cleaned_Dataset
```
